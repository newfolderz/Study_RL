### Lý thuyết xấp xỉ ngẫu nhiên 

**Động lực**: Tìm nghiệm của một hàm số hoặc điểm tối ưu khi không có công thức chính xác của hàm đó mà chỉ có thể quan sát được các giá trị bị nhiễu (noise)

*Thuật toán Robbins-Monro*: 

Vấn đề của 2 ông: điều chỉnh mức nhiệt độ của một phản ứng hóa học. Muốn tìm mức nhiệt $x$ sao cho hiệu suất đạt chính xác 90%. Vấn đề: mỗi lần đo thì kết quả lại bị sai lệch do sai số thiết bị hoặc tác động môi trường, cần một thuật toán tự điều chỉnh dựa trên các kết quả đo được. 

Giải phương trình: 
$$M(x) = a$$
Công thức cập nhật: 
$$x_{n+1} = x_n + a_n(\alpha - Y_n)$$
trong đó: 
- $x_n$: giá trị ước lượng tại bước thứ $n$
- $Y_n$: giá trị quan sát được tại bước n (đã gồm nhiễu $Y_n = M(X_n) + \epsilon_n$)
- $a_n$: hệ số bước hoặc learning rate 

Điều kiện hội tụ: 
- $\sum^{\infty}_{n=1} a_n = \infty$: tổng bước đi phải đủ lớn để thuật toán có thể chạm tới nghiệm dù ở bất cứ đâu 
- $\sum^{\infty}_{n=1} a_n^2 <> \infty$: các bước đi phải nhỏ dần theo thời gian để triệt tiêu ảnh hưởng của nhiễu 

VD kinh điển: Chuỗi $a_n = \frac{1}{n}$

Biến thể phổ biến: **Kiefer-Wolfowitz**
- Mục đích: dùng để tìm cực trị của hàm số khi không biết đạo hàm trực tiếp 
- Cách thức: ước lượng đạo hàm bằng cách đo sự khác biệt giữa 2 điểm gần nhau $M(x+c)$ và $M(x-c)$ rồi cập nhật vị trí theo độ dốc của đạo hàm. 

Ứng dụng thực tế: 
- Machine Learning: Stochastic Gradient Descent (SGD)
- Xử lý tín hiệu: các bộ lọc thích nghi tự điều chỉnh để loại bỏ tiếng ồn 
- Điều khiển tự động: Hệ thống tự lái hoặc robot tự điều chỉnh quỹ đạo dựa trên dữ liệu cảm biến không hoàn hảo. 
- Kinh tế: mô hình hóa sự học hỏi của các tác nhân trong thị trường khi thông tin bị hạn chế
